```python
# Імпорт необхідних бібліотек
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Генерація даних
m = 100
X = 6 * np.random.rand(m, 1) - 5  # Генеруємо значення для X в діапазоні [-5, 1]
y = 0.7 * X ** 2 + X + 3 + np.random.randn(m, 1)  # Залежність y від X (нелінійна)
```

```python
# Побудова графіка даних
plt.scatter(X, y, color='blue', label='Дані')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Графік залежності y = f(X)')
plt.legend()
plt.show()
```

```python
# Перетворення ознак на поліноміальні (degree=2)
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
```

```python
# Лінійна регресія на поліноміальних ознаках
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
```

```python
# Виведення коефіцієнтів регресії
print("Коефіцієнт регресії (y = b0 + b1 * X + b2 * X^2):")
print(f"Перехоплення (інтерсепт): {lin_reg.intercept_}")
print(f"Коефіцієнти (b1, b2): {lin_reg.coef_}")
```

```python
# Побудова графіка з лінією регресії
X_range = np.linspace(-5, 1, 100).reshape(100, 1)
X_range_poly = poly_features.transform(X_range)
y_range = lin_reg.predict(X_range_poly)

plt.scatter(X, y, color='blue', label='Дані')
plt.plot(X_range, y_range, color='red', label='Модель регресії')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Поліноміальна регресія (Degree = 2)')
plt.legend()
plt.show()
```
