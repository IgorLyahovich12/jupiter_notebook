```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
diabetes = datasets.load_diabetes()
X = diabetes.data  # 10 ознак
y = diabetes.target  # цільова змінна (прогресування хвороби)
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.5, random_state=0)
```

```python
regr = linear_model.LinearRegression()
regr.fit(Xtrain, ytrain)
```

```python
ypred = regr.predict(Xtest)
```

```python
print("Коефіцієнти регресії:", regr.coef_)
print("Зміщення (intercept):", regr.intercept_)
print("R2 score:", r2_score(ytest, ypred))
print("MAE:", mean_absolute_error(ytest, ypred))
print("MSE:", mean_squared_error(ytest, ypred))
```

```python
fig, ax = plt.subplots()
ax.scatter(ytest, ypred, edgecolors=(0, 0, 0))
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Виміряно')
ax.set_ylabel('Передбачено')
plt.title("Лінійна регресія: Прогноз vs Реальність")
plt.grid()
plt.show()
```
